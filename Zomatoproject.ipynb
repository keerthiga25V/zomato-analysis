```python
import pandas as pd

df = pd.read_csv('Zomato-data-.csv' , encoding='latin-1')
print(df.head())
print(df.shape)
```

```python
print(df.isnull().sum())
print(df.dtypes)
```

```python
print(df)
```

```python
# Step 1: "/5" part remove
# "4.1/5" → "4.1"
df['rate'] = df['rate'].str.replace('/5', '')

# Step 2: Extra spaces remove
# " 4.1 " → "4.1"
df['rate'] = df['rate'].str.strip()

# Step 3: String → Number convert
# "4.1" → 4.1
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

print(df['rate'].head())
```

```python
print(df['rate'].dtype)
```

```python
df.nlargest(10, 'votes')[['name', 'votes']]
```

```python
print(df['online_order'].value_counts())
```

```python
print(df.groupby('listed_in(type)')['rate'].mean().sort_values(ascending=False))
```

```python
print(df.groupby('approx_cost(for two people)')['name'].count())
```
